- Data aggrigating and grouping

- Mirging two dataframes


```python
df1 = pd.DataFrame({'key':['A','B','C','D'],'Value1':[1,2,3,8]})
df2 = pd.DataFrame({'key':['A','B','C','F'],'Value2':[4,5,6,7]})
```

```python
df1
```

```python
df2
```

- Merging two dataframes on the 'key' columns

- inner is for union intersection operation(common terms)

```python
pd.merge(df1,df2, on='key',how='inner')
```

- outer is for union operation(all the unique terms)

```python
pd.merge(df1,df2, on='key',how='outer')
```

```python
pd.merge(df1,df2, on='key',how='left')
```

```python
pd.merge(df1,df2, on='key',how='right')
```
